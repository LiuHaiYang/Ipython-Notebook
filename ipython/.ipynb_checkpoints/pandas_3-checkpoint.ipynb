```python
tuples = list(zip(*[['bar','bar','baz','baz',
               'foo','foo','qux','qux'],
               ['one','two','one','two',
               'one','two','one','two']]
                
))
```

```python
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples,names=['first','second'])
index
```

```python
df = pd.DataFrame(np.random.randn(8,2), index= index, columns=['A','B'] )
```

```python
df
```

```python
stacked = df.stack()
stacked
```

```python
stacked.index
```

```python
stacked.unstack
```

```python
stacked.unstack().unstack()
```
