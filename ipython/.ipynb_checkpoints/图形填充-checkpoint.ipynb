```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5*np.pi,1000)
y1 = np.sin(x)
y2 = np.sin(2*x) 
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.fill(x,y1,'b',alpha=0.3)
plt.fill(x,y2,'r',alpha=0.3)
```

```python
plt.plot(x,y1,x,y2,color='black')
plt.fill_between(x,y1,y2,facecolor='g')
```

```python
plt.plot(x,y1,color='r')
plt.plot(x,y2,color='g')
plt.fill_between(x,y1,y2,facecolor='yellow')
```

```python
plt.plot(x,y1,color='r')
plt.plot(x,y2,color='g')
plt.fill_between(x,y1,y2,where=y1>=y2,facecolor='yellow',interpolate=True)
plt.fill_between(x,y1,y2,where=y2>y1,facecolor='g',interpolate=True)
```
