```python
#我们再插入一个prop列，记录指定婴儿数量相对于总出生人数的比例，
#我们将按year&sex分组以后进行累加计算比例后新增加到每个分组中
def add_prop(group):
    births = group.births.astype(float) 
    group['prop']=births /births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
def get_top1k(group):
    return group.sort_index(by = 'births', ascending = False)
```

```python
grouped = names.groupby(['year', 'sex'])
```

```python
top1k = grouped.apply(get_top1k)
```

```python
top1k
```

```python
#分析命名的趋势
boys = top1k[top1k.sex == 'M']
girls = top1k[top1k.sex == 'F']
```

```python
#按照year & name 统计的总出生数透视表格
name7 = top1k.pivot_table('births', columns= 'name', index = 'year', aggfunc = sum)
```

```python
name7
```

```python
#利用DataFrame 的plot方法绘制几个名字的曲线图
subset = name7[['John', 'Harry', 'Mary','Emma' ]]
subset.plot(subplots=True, figsize = (12,10), grid= False, title='Number of births per year')
```
