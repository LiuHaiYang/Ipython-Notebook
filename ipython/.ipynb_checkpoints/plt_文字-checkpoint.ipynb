```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,11,1)
```

```python
y = x*x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.text(1,10,'ok')
```

```python
plt.plot(x,y)
plt.text(-2,10,'function:y=x*x',family='sans-serif',size=20,style='italic',weight=0)
plt.text(-2,20,'function:y=x*x',family='fantasy',size=15,color='g',style='italic',weight=1000)
plt.text(-2,40,'function:y=x*x',family='serif',size=20,color='y',style='italic',weight='black',bbox=dict(facecolor='r',alpha=0.5))
```
