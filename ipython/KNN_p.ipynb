```python
%matplotlib inline
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.colors import ListedColormap
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=12)
fontt=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=16)
```

```python
data= pd.read_csv('CCPP.csv',index_col='name')
```

```python
data
```

```python
virginica=data.ix['Iris-virginica']
setosa=data.ix['Iris-setosa']
versicolor=data.ix['Iris-versicolor']
```

```python
plt.scatter(setosa['hbl'],setosa['hbk'],label='setosa',c='r')
plt.scatter(versicolor['hbl'],versicolor['hbk'],label='versicolor',c='y')
plt.scatter(virginica['hbl'],virginica['hbk'],label='virginica',c='g')
plt.legend()
plt.title(u'花瓣',fontproperties=fontt)
plt.figtext(0.4,-0.01,'花瓣长度',fontproperties=font)
plt.figtext(-0.02,0.5,u'花瓣宽度',fontproperties=font)
```

```python
plt.scatter(setosa['epl'],setosa['epk'],label='setosa',c='r')
plt.scatter(versicolor['epl'],versicolor['epk'],label='versicolor',c='y')
plt.scatter(virginica['epl'],virginica['epk'],label='virginica',c='g')
plt.legend(loc=2)
plt.title(u'萼片',fontproperties=fontt)
plt.figtext(0.5,-0.02,u'萼片长度',fontproperties=font)
plt.figtext(-0.01,0.5,u'萼片宽度',fontproperties=font)
```
