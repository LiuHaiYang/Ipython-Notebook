```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.arange(1,5)
```

```python
plt.plot(y,y*2)
```

```python
plt.grid(True)
plt.plot(y,y*2)
```

```python
plt.grid(color='r')
plt.plot(y,y*2)
plt.grid(linewidth='2')
```

```python
plt.grid(color='r')
plt.plot(y,y*2)
plt.grid(linewidth='2')
plt.grid(linestyle='-')
```

```python
plt.grid(color='y')
plt.plot(y,y*2)
plt.grid(linewidth='2')
plt.grid(linestyle='-.')
```

```python
plt.grid(color='g')
plt.plot(y,y*2)
plt.grid(linewidth='2')
plt.grid(linestyle='--')
```

```python
plt.grid(color='b')
plt.plot(y,y*2)
plt.grid(linewidth='2')
plt.grid(linestyle='-.')
```
