```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
ts = pd.Series(np.random.rand(1000),index=pd.date_range('2000/1/1',periods=1000))
ts = ts.cumsum()
ts.describe()
```

```python
ts.plot(title='cumsum',style='r-',figsize=(8,6));
```

```python
df = pd.DataFrame(np.random.rand(1000,4),index=ts.index,columns=list('ABCD'))
df = df.cumsum()
df.describe()
```

```python
df.plot()
```

```python
df.plot(subplots=True,figsize=(6,12),sharey=True)
```
