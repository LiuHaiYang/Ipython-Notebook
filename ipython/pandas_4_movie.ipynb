```python
#电影的评分数据  100万条数据  6000  user  4000movie
#导入pandas
import pandas as pd
```

```python
#读取用户数据
#确定数据列名
unames = ['user_id','gender','age','occupation','zip']
#参数：文件名，分隔符，表头为空，设置列的名称
users = pd.read_table('ml-1m/users.dat',sep='::',header=None,names=unames)
#警告   read_table 数据解析器有两个版本，  c  和 python 这里用的是python
```

```python
print(len(users))
users.head(10) #默认打印5条
```

```python
#形同的方法导入电影和评分数据。
rating_names = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('ml-1m/ratings.dat',sep='::',header=None,names=rating_names)
```

```python
movie_names = ['movie_id','title','genres']
movies  = pd.read_table('ml-1m/movies.dat',sep='::',header=None,names=movie_names)
```

```python
#评分数据
print(len(ratings))
ratings.head()
```

```python
print(len(movies))
movies.head()
```

```python
#数据合并 通过外键 一一对应  调用merge函数  两次 
data = pd.merge(pd.merge(users,ratings),movies)
```

```python
print(len(data))
data
```

```python
data.head(10)
```

```python
#查看id = 1 用户的评分数据
data[data.user_id == 1]
```

```python
#所有电影不同性别的评分   数据透视pivot_table
ratings_by_gender = data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
```

```python
ratings_by_gender.head(10)
```

```python
#求出男女评分差异最大的电影，评分相减      新加一个列 diff
ratings_by_gender['diff'] = ratings_by_gender.F - ratings_by_gender.M
```

```python
ratings_by_gender.head(10)
```

```python
#对 diff 进行排序  升序
ratings_by_gender.sort_values(by='diff', ascending=True).head(10)
```

```python
#降序  
ratings_by_gender.sort_values(by='diff', ascending=False).head(10)

#  男女之间的分歧
```

```python
#title进行分组。 size 评分人数
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title.head(10)
```

```python
#排序  倒序 
ratings_by_title.sort_values(ascending=False).head(10)
```

```python

ratings_by_title.sort_values(ascending=True).head(10)
```

```python
#每部电影的平均得分
mean_ratings = data.pivot_table(values='rating',index='title',aggfunc='mean')
```

```python
mean_ratings.head(10)
```

```python
#评分最低
mean_ratings.sort_values(ascending=True).head(10)
```

```python
#评分最高
mean_ratings.sort_values(ascending=False).head(10)
```

```python
#之前找的前十大热门电影 评分人数
top_hot_10 = ratings_by_title.sort_values(ascending=False).head(10)
top_hot_10
```

```python
#平均的评分数据（前十大）
mean_ratings[top_hot_10.index]
```

```python
#平均评分数据（前20大）
top_20_score = mean_ratings.sort_values(ascending=False).head(20)
```

```python
top_20_score
```

```python
#查看前20高分电影的 评分人数
ratings_by_title[top_20_score.index]
```
