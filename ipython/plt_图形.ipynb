```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.patches as mpatches
```

```python
fig, ax =plt.subplots()
```

```python
fig, ax=plt.subplots()
xyl = np.array([0.2,0.2])
circle = mpatches.Circle(xyl,0.1)
ax.add_patch(circle)
```

```python
fig, ax=plt.subplots()
xyl = np.array([0.2,0.2])
xy2 = np.array([0.1,0.8])
circle = mpatches.Circle(xyl,0.05)
rectangle = mpatches.Rectangle(xy2,0.2,0.1,color='r')
plt.grid()
ax.add_patch(circle)
ax.add_patch(rectangle)
```
